#ランベルトの多項式展開を有理多項式にするための係数計算

ランベルトの多項式を指定した項数 (n)までで切ったものを下から順に計算していくと、分母に偶数次、分子に奇数次の項が残る。このときの各項の係数を計算するプログラムです。

本編：[ランベルト連分数から有理関数への導出と n=13 の全係数](https://tukumolog.com/lambert-tan-coefficients-n13/)

```python
import sympy as sp

def get_lambert_coeffs(n_target):
    x = sp.Symbol('x')

    p = [sp.Integer(0), x]
    q = [sp.Integer(1), sp.Integer(1)]

    x_sq = x**2

    for n in range(2, n_target + 1):
        coeff = 2 * n - 1
        # 漸化式：P_n = (2n-1)P_{n-1} - x^2 P_{n-2}
        new_p = sp.expand(coeff * p[n-1] - x_sq * p[n-2])
        new_q = sp.expand(coeff * q[n-1] - x_sq * q[n-2])

        p.append(new_p)
        q.append(new_q)

    return p[n_target], q[n_target], x
```

```python
n = 13
# nまで計算
P_n, Q_n, x = get_lambert_coeffs(n)

# 比較用に整形して表示
print(f"{'次数 k':>5} | {'分子 P(x) の係数':>25} | {'分母 Q(x) の係数':>25}")
print("-" * 65)

for k in range(n+1):
    pk = P_n.coeff(x, k)
    qk = Q_n.coeff(x, k) if k <= 12 else 0
    print(f"x^{k:2d} | {int(pk):25,d} | {int(qk):25,d}")
```
